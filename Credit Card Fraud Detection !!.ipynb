### Credit Card Fraud Detection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Seperating the data for analysis.

```python
legit_data = dataset[dataset.Class == 0]
fraud_data = dataset[dataset.Class == 1]
```

```python
legit_data.shape
```

```python
fraud_data.shape
```

```python
legit_data.Amount.describe()
```

```python
fraud_data.Amount.describe()
```

```python
dataset.groupby('Class').mean()
```

The dataset is unbalanced, to sovle, we use Undersampling in which we have to build a dataset containing similar distribution of Legit and Fraud transactions.

```python
sample_legit = legit_data.sample(n=492)
```

### Concatenate sample_legit and fraud_data

```python
new_data = pd.concat([sample_legit, fraud_data], axis = 0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data.Class.value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
X = new_data.drop(columns = 'Class', axis = 1)
Y = new_data.Class
```

```python
X.head()
```

### Split the data into train and test data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### Training the model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

### Model Evaluation based on Accuracy score

```python
from sklearn.metrics import accuracy_score
```

```python
X_train_pred = model.predict(X_train)
training_accuracy = accuracy_score(X_train_pred, Y_train)
```

```python
print('Accuracy on training data:',training_accuracy)
```

```python
X_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(X_test_pred, Y_test)
```

```python
print('Accuracy on test data:',test_accuracy)
```
